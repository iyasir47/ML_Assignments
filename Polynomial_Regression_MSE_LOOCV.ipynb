To make the 3rd task a bit easier, here is reminder on how to plot your data and a polynomial:

```python
lin_reg = LinearRegression()

# preparing the model
poly_features = PolynomialFeatures(degree=2, include_bias=False)
X_poly = poly_features.fit_transform(X)  # adding polynomial features
lin_reg.fit(X_poly, y)  # fitting the model to the data

# preparing the x,y values for plotting
X_plot = np.linspace(np.min(X), np.max(X), 100).reshape(100, 1)  # values at which to predict y
poly_features = PolynomialFeatures(degree=2, include_bias=False)
X_poly = poly_features.fit_transform(X_plot)  # adding polynomial features
y_plot = lin_reg.predict(X_poly)  # computing predictions which will be used for plotting the curve

# actual plotting
plt.figure(figsize=(6, 4))
plt.plot(X, y, 'b.', label='Data')
plt.plot(X_plot, y_plot, label='Model')
plt.xlabel('$x_1$', fontsize=14)
plt.ylabel('$y$', fontsize=14)
plt.legend(loc='upper left')
plt.grid()
plt.show()
```
